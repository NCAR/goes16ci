Compare Epoch Loss with Other Valuable Metrics

```python
import yaml
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_path = "benchmark_output/benchmark_output_final_"
image_size = ["32","64"]
epoch_num = ["1","5","10","15","20","30"]
epoch_loss_list = []
for i in image_size:
    for j in epoch_num:
        path = data_path + i + "_" + j + "/goes_benchmark_data.yml"
        with open(path,"r") as file:
            bench_output = yaml.load(file, Loader = yaml.Loader)
        epoch_loss = bench_output["epoch_loss"]
        epoch_loss_list.append([i,j,epoch_loss])
print(epoch_loss_list)    
```

```python
epoch_32 = []
epoch_64 = []
epoch_loss_32 = []
epoch_loss_64 = []
for x in epoch_loss_list:
    if (x[0] == '32'):
        epoch_32.append(x[1])
        epoch_loss_32.append(x[2])
    elif (x[0] == '64'):
        epoch_64.append(x[1])
        epoch_loss_64.append(x[2])
print(epoch_32)
print(epoch_64)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
i = 32
j = 64
#blue
i = ax.scatter(epoch_32,epoch_loss_32)
i = ax.plot(epoch_32,epoch_loss_32)
#orange
j = ax.scatter(epoch_64,epoch_loss_64)
j = ax.plot(epoch_64,epoch_loss_64)
plt.title('Epoch Loss Dependent on Number of Epochs')
plt.xlabel('Number of Epochs')
plt.ylabel('Epoch Loss')
#plt.legend()
plt.show()
```
